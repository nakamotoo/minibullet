```python
import roboverse
import numpy as np
from PIL import Image
def _process_image(obs):
    obs = (obs * 255).astype(np.uint8)
    obs = np.reshape(obs, (3, 128, 128))
    return np.transpose(obs, (1, 2, 0))
```

```python
env_name = 'Putcolunnade_topintoBowlDiverseTwoDistractors-v0'
# env_name = 'Putglass_half_gallonintoBowlDiverse-v0'
# env_name = 'PutT_cupintoBowlDiverse-v0'

# env_name = 'PutBallintoBowlTwoDistractors-v0'
# env_name = 'Widow250PickPlaceMultiObjectMultiContainerTrain-v0'
env = roboverse.make(env_name, transpose_image=True)
```

```python
obs = env.reset()
obs  = _process_image(obs["image"])
obs = Image.fromarray(obs)
obs
```
